# Seaborn Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips').head(20)
df
```

### 1. Line Plot

```python
sns.lineplot(x='total_bill', y='tip', data=df, hue='sex',style='sex', markers=True, #hue draws another line based on no. of categories
             palette='tab10', linewidth=2 )                                             #style changes style of each line based on category

plt.grid()
plt.show()
```

### 2. Histogram Plot

```python
# sns.displot(x='tip', data=df)
sns.displot(df['total_bill'], kde=True, rug=True)     #bins=[10,15,20,25,30,35] for equal sized bins

plt.show()                                            #rug for scale, kde for curve
```

### 3. Bar Plot

```python

tips = sns.load_dataset('tips')

sns.barplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    hue_order=['Female', 'Male'],
    orient='v',
    err_kws={'color': 'grey'},
)

sns.set(style='darkgrid')
plt.show()
```

### 4. Scatter Plot

```python
m={'Male':'o', 'Female':'X'}

sns.scatterplot(
x='total_bill',
y= 'tip',
data= df,
hue= 'sex',
style= 'sex',
size='sex',
sizes=(100,50),   #female legend size, male size
palette='binary',
markers=m    #custom markers (m is dictionary)
)

plt.show()
```

### 5. Heatmap Plot

```python
data = np.random.rand(10, 5)

sns.heatmap(
    data,
    annot=True,           # to annotate cells with values
    fmt=".1f",           # to change format of annotated value to float
    cmap="viridis",
    linewidths=0.5,      # to alter width of line between cells
    linecolor='black',
    cbar=True,           # to display color bar
)

plt.title('Heatmap of Random Values')
plt.show()
```
