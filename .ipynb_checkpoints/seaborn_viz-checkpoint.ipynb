# Seaborn Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('tips').head(20)
dataset
```

### 1. Line Plot

```python
sns.lineplot(x='total_bill', y='tip', data=dataset, hue='sex',style='sex', markers=True, #hue draws another line based on no. of categories
             palette='tab10', linewidth=2 )                                             #style changes style of each line based on category
plt.grid()
plt.show()
```
